```python
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
```

```python
ER = nx.erdos_renyi_graph(n=300, p=0.1)
ne = np.random.randint(low=1, high=(10*(10-1))/2)
RG = nx.dense_gnm_random_graph(n=10, m=ne)
```

```python
nx.draw(ER)
```

```python
print(RG, np.random.choice(np.array(RG.nodes())))
```

```python
for i in range(290):
    RG.add_node(i+300)
    for _ in range(3):
        node = np.random.choice(np.array(RG.nodes()))
        node_degree = RG.degree[node]
        sigma_degrees = np.sum([val for _,val in RG.degree()])
        p = node_degree/sigma_degrees
        if np.random.rand() <= p:
            RG.add_edge(i+300, node)
```

```python
RG_degree = np.zeros((300,1))
ER_degree = np.zeros((300,1))
for _, val in RG.degree:
    RG_degree[val]+=1
for _, val in ER.degree():
    ER_degree[val]+=1
```
